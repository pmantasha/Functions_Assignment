1. What is the difference between a function and a method in Python?
 - Function: In python function is a block of code that performs a specific task and can be called independently. It is defined using the def keyword and can be called without an object.

 def greet():

2.  Explain the concept of function arguments and parameters in Python.

 def greet(name):  # 'name' is the parameter

5. What are iterators in Python and how do they differ from iterables?
 - Iterable: An iterable is an object that can be looped over (iterated) using a for loop or other iteration mechanisms. Examples include lists, tuples, sets, and dictionaries. An iterable must implement the __iter__() method or have a __getitem__() method.

 - Iterator: An iterator is an object that represents a stream of data. It is used to traverse through all the elements of an iterable one by one. An iterator must implement two methods: __iter__() (which returns the iterator itself) and __next__() (which returns the next item and raises StopIteration when there are no more items).

```python
#example:-
my_list = [1, 2, 3]  # Iterable
my_iter = iter(my_list)  # Iterator

print(next(my_iter))
print(next(my_iter))
```

6. Explain the concept of generators in Python and how they are defined.
 - Generators in Python are a type of iterable that allow for lazy evaluation. They generate values on the fly, one at a time, instead of storing all values in memory, which makes them memory-efficient.

 #Key Concepts:
 - Defined using yield: A generator is defined using a function that contains the yield keyword. Each time yield is called, the function pauses and returns the value to the caller. The state of the function is saved, and it resumes where it left off when called again.

 - Lazy Evaluation: Generators compute values only when needed, making them more efficient for large datasets or infinite sequences.

 Difference from normal functions: Normal functions return a single value and terminate, while generators return an iterator and can yield multiple values, pausing and resuming each time next() is called.

```python
#example:-
def count_up_to(limit):
    count = 1
    while count <= limit:
        yield count  # Pauses and returns count
        count += 1

# Creating a generator object
counter = count_up_to(5)

# Accessing values one by one using next()
print(next(counter))
print(next(counter))
```

7. What are the advantages of using generators over regular functions?
 - Advantages of using generators over regular functions:

  a. Memory Efficiency: Generators generate values on-the-fly (lazily) instead of storing them all in memory, making them more memory-efficient, especially for large datasets or infinite sequences.

  b. Performance: Since generators yield values one at a time, they can improve performance by reducing the overhead of creating and storing large collections in memory.

  c. Lazy Evaluation: Generators only compute values when requested, which can lead to faster execution when dealing with large inputs or complex computations.

  d. Convenience: Generators provide a simple way to implement iterators without needing to manually manage state or memory.

  In summary, generators are more efficient for handling large or infinite data streams and can be more performant compared to regular functions that return entire collections.

9.  Explain the purpose and usage of the `map()` function in Python.
 - The map() function in Python is used to apply a given function to all items in an iterable (such as a list or tuple) and returns an iterator (map object) of the results.

   Syntax: map(function, iterable)

  a. function: A function to apply to each item in the iterable.
   
  b. iterable: An iterable (e.g., list, tuple) whose items are passed to the function.

 - Usage:

  a. Transforming data: Itâ€™s typically used when you want to apply a transformation or operation to each element of an iterable.

  b. Combining with lambda: Often used with lambda functions for short, one-time operations.


```python
#example:-
numbers = [1, 2, 3, 4]
squared = map(lambda x: x ** 2, numbers)  # Squaring each number
print(list(squared))
```

10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?
 - In Python, map(), reduce(), and filter() are higher-order functions used for processing iterables, but they serve different purposes:

  a. map():

   Purpose: Applies a function to all items in an iterable and returns a new iterable (map object) with the results.
   
   Usage: Transforms each element of the iterable.

```python
#example:-
numbers = [10, 20, 30]
result = map(lambda x: x * 2, numbers)
print(list(result))
```

   b. reduce() (from the functools module):

  Purpose: Applies a binary function (takes two arguments) cumulatively to the items of an iterable, reducing the iterable to a single value.
  
  Usage: Used for aggregation or accumulation of results.


```python
#example:-
from functools import reduce
numbers = [1, 2, 3, 4]
result = reduce(lambda x, y: x + y, numbers)
print(result)
```

   c. filter():

  Purpose: Filters the elements of an iterable based on a condition defined by a function, returning a new iterable with elements that satisfy the condition.

  Usage: Used to select items that meet a criteria.

```python
#example:-
numbers = [1, 2, 3, 4, 5]
result = filter(lambda x: x % 2 == 0, numbers)
print(list(result))
```

```python
#answer:-
def sum_of_even_numbers(numbers):
    # Using list comprehension to filter even numbers and sum them
    return sum([num for num in numbers if num % 2 == 0])

# Example usage:
numbers = [1, 2, 3, 4, 5, 6, 7, 8]
result = sum_of_even_numbers(numbers)
print(result)
```

```python
#4. Write a Python function that checks if a given number is prime or not from 1 to 200.
```

```python
#answer:
def is_prime(num):
    # Handle cases for numbers less than 2
    if num <= 1:
        return False
    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            return False
    return True

# Example usage:
for number in range(1, 201):
    if is_prime(number):
        print(number)
```

```python
#5.  Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.
```

```python
#answer:-
class FibonacciIterator:
    def __init__(self, n):
        self.n = n
        self.a, self.b = 0, 1
        self.count = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.count < self.n:
            current = self.a
            self.a, self.b = self.b, self.a + self.b
            self.count += 1
            return current
        else:
            raise StopIteration

# Example usage:
fib = FibonacciIterator(10)
for num in fib:
    print(num)
```

```python
#6. Write a generator function in Python that yields the powers of 2 up to a given exponent.
```

```python
#answer:-
def powers_of_two(exponent):
    for i in range(exponent + 1):
        yield 2 ** i

# Example usage:
for power in powers_of_two(5):
    print(power)
```

```python
#7.  Implement a generator function that reads a file line by line and yields each line as a string.
```

```python
#answer:-
def read_file_line_by_line(file_path):

    try:
        with open(file_path, 'r') as file:
            for line in file:
                yield line.rstrip('\n')  # Strip newline characters for cleaner output
    except FileNotFoundError:
        print(f"The file at '{file_path}' was not found.")
    except Exception as e:
        print(f"An error occurred: {e}")
```

```python
#9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.
```

```python
#answer:-
# List of temperatures in Celsius
celsius_temps = [0, 25, 30, 37, 100]

# Function to convert Celsius to Fahrenheit
def celsius_to_fahrenheit(celsius):
    return (celsius * 9/5) + 32

# Using map() to apply the conversion function to each element in the list
fahrenheit_temps = list(map(celsius_to_fahrenheit, celsius_temps))

# Printing the converted temperatures
print(fahrenheit_temps)
```

```python
#answer:-
# Function to check if a character is not a vowel
def is_not_vowel(char):
    return char.lower() not in 'aeiou'

# Given string
input_string = "Hello, World!"

# Using filter() to remove vowels from the string
filtered_string = ''.join(filter(is_not_vowel, input_string))

# Print the result
print(filtered_string)
```
